```python
# 1. User Login Check
# Stored credentials
correct_username = "admin"
correct_password = "1234"

# Given credentials (you can change these to test)
username = "admin"
password = "1234"

# Check login
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Check Even or Odd
number = 10  # change this value to test

if number % 2 == 0:
    print("\nThe number is Even")
else:
    print("\nThe number is Odd")
```

```python
# 3. Find Largest of Three Numbers
a = 10
b = 25
c = 15

if a > b and a > c:
    largest = a
elif b > a and b > c:
    largest = b
else:
    largest = c

print("\nLargest number is:", largest)
```

```python
# 4. Student Grade System
marks = 78  # change this to test different grades

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 50:
    grade = "D"
else:
    grade = "Fail"

print("\nStudent Grade:", grade)
```

```python
# 5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR", "INFO", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nError Log Analysis:")
print("Total ERROR entries:", error_count)
```
